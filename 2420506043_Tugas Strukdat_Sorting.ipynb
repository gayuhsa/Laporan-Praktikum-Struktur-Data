```python
# Modul
import prettytable
import random

# Variabel
sorted_table = prettytable.PrettyTable(["No", "Nama", "Nilai"])
unsorted_table = prettytable.PrettyTable(["No", "Nama", "Nilai"])
students = []

# Class
class Student:
  def __init__(self, name, score):
    self.name = name
    self.score = score

# Fungsi
def quick_sort(array_to_sort):
  if len(array_to_sort) <= 1:
    return array_to_sort

  pivot = array_to_sort[len(array_to_sort) // 2]
  left = [x for x in array_to_sort if x.score > pivot.score]
  middle = [x for x in array_to_sort if x.score == pivot.score]
  right = [x for x in array_to_sort if x.score < pivot.score]

  return quick_sort(left) + middle + quick_sort(right)


# Program utama
for i in range(-1, 10):
  students.append(Student("E", random.randrange(0, 100)))

for i in range(len(students) - 1):
  unsorted_table.add_row([i + 1, students[i].name, students[i].score])

students = quick_sort(students.copy())

for i in range(len(students) - 1):
  sorted_table.add_row([i + 1, students[i].name, students[i].score])

print(f"Sebelum diurutkan:\n{unsorted_table}\n")
print(f"Setelah diurutkan:\n{sorted_table}")
```
