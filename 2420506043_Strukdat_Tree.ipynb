```python
# Class
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

class BinarySearchTree:
  def __init__(self):
    self.root = None

  def insert(self, root, data):
    if root is None:
      return Node(data)
    if data < root.data:
      root.left = self.insert(root.left, data)
    else:
      root.right = self.insert(root.right, data)
    return root


# Fungsi
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=" ")
    inorder(node.right)

def preorder(node):
  if node:
    print(node.data, end=" ")
    preorder(node.left)
    preorder(node.right)

def postorder(node):
  if node:
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=" ")

def search(root, target):
  if root is None or int(root.data) == target:
    return root
  if target < int(root.data):
    return search(root.left, target)
  else:
    return search(root.right, target)


# Program utama
bst = BinarySearchTree()
root = None
data_list = [
    int(input("Tiga digit terakhir NPM Anda: ")),
    int(input("Tanggal lahir Anda (DD): ")),
    int(input("Bulan lahir Anda (MM): ")),
    int(input("ASCII dari huruf pertama nama depan Anda: ")),
    int(input("ASCII dari huruf kedua nama depan Anda: ")),
    int(input("Angka bebas (1/2): ")),
    int(input("Angka bebas (2/2): "))
]

for value in data_list:
  root = bst.insert(root, value)

print("\nTraversal inorder:")
print(inorder(root))
print()

print("Traversal preorder:")
print(preorder(root))
print()

print("Traversal postorder:")
print(postorder(root))
print()


# Loop pencarian
while True:
  query = input("Cari: ")

  if query == "stop":
    break

  print(f"{search(root, int(query))}\n")
```
