```python
import random

possible_items = ["Apel", "Bawang merah", "Bawang putih", "Beras", "Cabai", "Daging ayam", "Garam", "Gula", "Ikan", "Jeruk", "Minyak", "Sayur", "Teh", "Telur"]
shopping_list = []

random.seed()

for i in range(random.randrange(len(possible_items))):
  # Hentikan kode program jika possible_items sudah habis
  if len(possible_items) <= 0:
    break

  new_item = possible_items[0]

  if len(possible_items) > 1:
    new_item = possible_items[random.randrange(0, len(possible_items) - 1)]

  # Hapus item yang terpilih dari possible_items agar tidak terpilih 2 kali
  possible_items.remove(new_item)

  shopping_list.append(new_item)

print(shopping_list)
```
