# Algoritma Sorting

```python
# Modul
import random

# Variabel
array = random.sample(range(1, 1000), 10)

# Fungsi
def bubble_sort(array_to_sort):
  n = len(array_to_sort)
  for i in range(n - 1):
    for j in range(n - i - 1):
      if array_to_sort[j] > array_to_sort[j + 1]:
        array_to_sort[j], array_to_sort[j + 1] = array_to_sort[j + 1], array_to_sort[j]

  return array_to_sort

def selection_sort(array_to_sort):
  n = len(array_to_sort)

  for i in range(n - 1):
    min_index = i
    for j in range(i + 1, n):
      if array_to_sort[j] < array_to_sort[min_index]:
        min_index = j
    min_value = array_to_sort.pop(min_index)
    array_to_sort.insert(i, min_value)

  return array_to_sort

def insertion_sort(array_to_sort):
  n = len(array_to_sort)

  for i in range(1, n):
    insert_index = i
    current_value = array_to_sort.pop(i)
    for j in range(i - 1, -1, -1):
      if array_to_sort[j] > current_value:
        insert_index = j
    array_to_sort.insert(insert_index, current_value)

  return array_to_sort

def merge_sort(array_to_sort):
  if len(array_to_sort) > 1:
    mid = len(array_to_sort) // 2
    left_half = array_to_sort[:mid]
    right_half = array_to_sort[mid:]

    merge_sort(left_half)
    merge_sort(right_half)

    i = j = k = 0

    while i < len(left_half) and j < len(right_half):
      if left_half[i] < right_half[j]:
        array_to_sort[k] = left_half[i]
        i += 1
      else:
        array_to_sort[k] = right_half[j]
        j += 1
      k += 1

    while i < len(left_half):
      array_to_sort[k] = left_half[i]
      i += 1
      k += 1

    while j < len(right_half):
      array_to_sort[k] = right_half[j]
      j += 1
      k += 1

  return array_to_sort

def quick_sort(array_to_sort):
  if len(array_to_sort) <= 1:
    return array_to_sort

  pivot = array_to_sort[len(array_to_sort) // 2]
  left = [x for x in array_to_sort if x < pivot]
  middle = [x for x in array_to_sort if x == pivot]
  right = [x for x in array_to_sort if x > pivot]

  return quick_sort(left) + middle + quick_sort(right)


print(f"Array sebelum diurutkan:\n{array}\n")
print(f"Array setelah diurutkan dengan bubble sort:\n{bubble_sort(array.copy())}\n")
print(f"Array setelah diurutkan dengan selection sort:\n{selection_sort(array.copy())}\n")
print(f"Array setelah diurutkan dengan insertion sort:\n{insertion_sort(array.copy())}\n")
print(f"Array setelah diurutkan dengan merge sort:\n{merge_sort(array.copy())}\n")
print(f"Array setelah diurutkan dengan quick sort:\n{quick_sort(array.copy())}\n")
```
