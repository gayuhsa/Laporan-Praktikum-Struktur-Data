```python
import matplotlib
import networkx

edges = []
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}
network_graph = networkx.Graph()


for key in list(graph.keys()):
  temp = []

  for v in graph[key]:
    temp = [key]
    temp.append(v)
    edges.append(tuple(temp))
    temp = []


network_graph.add_edges_from(edges)
networkx.draw(network_graph, with_labels=True)
matplotlib.pyplot.show()
```

```python
# Modul
from collections import deque


# Variabel
test_graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}


# Fungsi
def bfs(graph, start):
  visited = set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end=" ")
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])


def dfs(graph, start, visited=None):
  if visited is None:
    visited = set()

  if start not in visited:
    print(start, end=" ")
    visited.add(start)
    for neighbor in graph[start]:
      dfs(graph, neighbor, visited)


# Program Utama
print("BFS:")
print(bfs(test_graph, "A"))
print("\nDFS:")
print(dfs(test_graph, "A"))
```

```python
# Modul
from collections import deque


# Variabel
test_graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}


# Fungsi
def find_path(graph, start, finish, visited=None, path=None, is_found=None):
  if visited is None:
    visited = set()
    is_found = [False]
    path = []

  if is_found[0]:
    return

  if start not in visited:
    path.append(start)
    visited.add(start)

    if start == finish:
      is_found[0] = True
      return path

    for neighbor in graph[start]:
      find_path(graph, neighbor, finish, visited, path, is_found)
      if is_found[0]:
        return path

    if not is_found[0]:
      path.pop()


# Program Utama
print("Path:")
print(find_path(test_graph, "A", "D"))
```

```python
# Modul
from collections import deque


# Variabel
test_graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E'],
}


# Fungsi
def is_connected(graph, start=None, visited=None):
  if start is None:
    start = list(graph.keys())[0]

  if visited is None:
    visited = set()

  if start not in visited:
    visited.add(start)
    for neighbor in graph[start]:
      is_connected(graph, neighbor, visited)

  return len(visited) == len(list(graph.keys()))


# Program Utama
print(f"is_connected(): {is_connected(test_graph)}")
```
