```python
# Class
class Node:
  def __init__(self, data):
    self.data = data
    self.next_node = None

class SinglyLinkedList:
  def __init__(self):
    self.head = None

  # Fungsi iterator untuk linked list
  def  __iter__ (self):
    node = self.head

    while node is not None:
      yield node
      node = node.next_node

  def at(self, index):
    current_index = 0

    for v in self:
      if current_index == index:
        return v.data
      current_index = current_index + 1

    return None

  def length(self):
    # Linked list kosong, langsung return 0
    if self.head == None:
      return 0

    length = 0

    for _ in self:
      length = length + 1

    return length

  def display(self):
    elements = []

    for v in self:
      elements.append(v.data)

    print(" -> ".join(elements))

  def find_node(self, target):
    for v in self:
      if v.data == target:
        return v

    return None

  def insert(self, new_data, add_index = None):
    # Linked list kosong, buat node baru sebagai head
    if not self.head:
      self.head = Node(new_data)
      return

    if isinstance(add_index, int): # Sisipkan elemen pada add_index
      current_index = 0
      previous_node = None
      for v in self:
        if current_index == add_index:
          new_node = Node(new_data)
          if v == self.head: # Sisipkan elemen di awal
            self.head = new_node
            new_node.next_node = v
          else:
            new_node.next_node = v
            previous_node.next_node = new_node
          break
        current_index = current_index + 1
        previous_node = v
    else: # Tambahkan elemen di akhir
      last_element = self.head
      for v in self:
        last_element = v
      last_element.next_node = Node(new_data)

  def pop(self, pop_index = None):
    current_index = 0
    previous_node = None

    for v in self:
      # Jika pop_index adalah integer, hapus elemen pada pop_index
      # Jika pop_index adalah None, hapus elemen terakhir
      if (isinstance(pop_index, int) and current_index == pop_index) or (pop_index == None and v.next_node == None):
        if v == self.head:
          self.head = v.next_node
        else:
          previous_node.next_node = v.next_node
        break
      current_index = current_index + 1
      previous_node = v


schedule_linked_list = SinglyLinkedList()

for subject in ["Praktikum Struktur Data", "Praktikum Basis Data", "Fisika Lanjut"]:
  schedule_linked_list.insert(subject)

schedule_linked_list.display()
```

```python
# Class
class DoublyNode:
  def __init__(self, data):
    self.data = data
    self.next_node = None
    self.previous_node = None

class DoublyLinkedList:
  def __init__(self):
    self.current = None
    self.head = None

  # Fungsi iterator untuk linked list
  def  __iter__ (self):
    node = self.head

    while node is not None:
      yield node
      node = node.next_node

  def display(self):
    elements = []

    for v in self:
      elements.append(v.data)

    print(" <-> ".join(elements))

  def go_backward(self):
    current = self.current
    if current and current.previous_node is not None:
      self.current = current.previous_node
      print(f"Mundur ke {self.current.data}")
    else:
      print("Elemen pertama. Tidak bisa mundur")

  def go_forward(self):
    current = self.current
    if current and current.next_node is not None:
      self.current = current.next_node
      print(f"Maju ke {self.current.data}")
    else:
      print("Elemen terakhir. Tidak bisa maju")

  def insert(self, new_data, add_index = None):
    # Linked list kosong, buat node baru sebagai head
    if not self.head:
      self.head = DoublyNode(new_data)
      self.current = self.head
      return

    if isinstance(add_index, int): # Sisipkan elemen pada add_index
      current_index = 0
      prev = None
      for v in self:
        if current_index == add_index:
          new_node = DoublyNode(new_data)
          if v == self.head: # Sisipkan elemen di awal
            self.head = new_node
            new_node.next_node = v
            v.previous_node = new_node
          else:
            new_node.next_node = v
            new_node.previous_node = prev
            v.previous_node = new_node
            prev.next_node = new_node
          break
        current_index = current_index + 1
        prev = v
    else: # Tambahkan elemen di akhir
      new_node = DoublyNode(new_data)
      last_element = self.head
      for v in self:
        last_element = v
      last_element.next_node = new_node
      new_node.previous_node = last_element

  def pop(self, pop_index = None):
    current_index = 0
    previous_node = None

    for v in self:
      # Jika pop_index adalah integer, hapus elemen pada pop_index
      # Jika pop_index adalah None, hapus elemen terakhir
      if (isinstance(pop_index, int) and current_index == pop_index) or (pop_index == None and v.next_node == None):
        if v == self.head:
          self.head = v.next_node
          self.head.previous_node = None
        else:
          previous_node.next_node = v.next_node
          v.next_node.previous_node = previous_node
        break
      current_index = current_index + 1
      previous_node = v


history = DoublyLinkedList()

for directory in ["C://Program Files", "C://Program Files (x86)", "E://Endless Sky", "E://Paint.NET"]:
  history.insert(directory)

history.display()
print(f"\nMulai di {history.current.data}\n")

for _ in range(4):
  history.go_forward()

print(f"\nSaat ini di {history.current.data}")

for _ in range(4):
  history.go_backward()
```
